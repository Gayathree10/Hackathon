```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Ladder score'], kde=True)
plt.title('Distribution of Ladder Score')
plt.xlabel('Ladder Score')
plt.ylabel('Frequency')
plt.show()
```

```python
correlation_matrix = df.corr()
correlation_matrix
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Logged GDP per capita', y='Ladder score', data=df)
plt.title('Logged GDP per capita vs Ladder Score')
plt.xlabel('Logged GDP per capita')
plt.ylabel('Ladder Score')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Social support', y='Ladder score', data=df)
plt.title('Social support vs Ladder Score')
plt.xlabel('social support')
plt.ylabel('Ladder Score')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Healthy life expectancy', y='Ladder score', data=df)
plt.title('Healthy life expectancy vs Ladder Score')
plt.xlabel('Healthy life expectancy')
plt.ylabel('Ladder Score')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Freedom to make life choices', y='Ladder score', data=df)
plt.title('Freedom to make life choices vs Ladder Score')
plt.xlabel('Freedom to make life choices')
plt.ylabel('Ladder Score')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Generosity', y='Ladder score', data=df)
plt.title('Generosity vs Ladder Score')
plt.xlabel('Generosity')
plt.ylabel('Ladder Score')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Perceptions of corruption', y='Ladder score', data=df)
plt.title('Perceptions of corruption vs Ladder Score')
plt.xlabel('Perceptions of corruption')
plt.ylabel('Ladder Score')
plt.show()
```

```python
top_bottom_10 = df.nlargest(10, 'Ladder score')[['Country name', 'Ladder score']]
top_bottom_10 = top_bottom_10.append(df.nsmallest(10, 'Ladder score')[['Country name', 'Ladder score']])
top_bottom_10
```

```python
plt.figure(figsize=(15, 10))
sns.boxplot(x='Regional indicator', y='Ladder score', data=df)
plt.title('Ladder Score Distribution by Regional Indicator')
plt.xlabel('Regional Indicator')
plt.ylabel('Ladder Score')
plt.xticks(rotation=45)
plt.show()
```

```python
import plotly.express as px

fig = px.choropleth(
    df,
    locations="Country name",
    locationmode='country names',
    color="Ladder score",
    hover_name="Country name",
    projection="natural earth",
    title='Ladder Score Distribution by Countries (Map View)',
)
fig.show()
```

```python
most_generous_2021 = df.sort_values(by='Generosity', ascending=False).head(10)
least_generous_2021 = df.sort_values(by='Generosity', ascending=True).head(10)
print("Most Generous Countries:")
print(most_generous_2021[['Country name', 'Generosity']])
print("\nLeast Generous Countries:")
print(least_generous_2021[['Country name', 'Generosity']])
```
